```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/op1154/BaseDadosImoveisRJ/main/aluguel.csv"

df = pd.read_csv(url, sep=";")

df.head()
```

```python
df['Valor'].mean()
```

```python
df.dtypes
```

```python
df.groupby('Tipo')
```

```python
df.groupby('Tipo').mean(numeric_only=True)
```

```python
df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_preco_tipo = df.groupby('Tipo').mean(numeric_only=True)[['Valor']].sort_values("Valor")
```

```python
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color ='purple');

plt.show()
```

```python
df_preco_tipo.plot(kind='line', marker='o')
```

```python
plt.title('Média do Valor de Alugueis por tipo de imovel no RJ')
plt.ylabel('Tipo Imovel')
plt.grid(True)
plt.show()
```

```python
df_imoveis.Tipo.value_counts()
```

```python
df_imoveis.Tipo.value_counts(normalize=True)
df_imoveis.Tipo.value_counts(normalize=True).to_frame()
df_imoveis['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo = df_imoveis['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color ='green', edgecolor='black', xlabel = 'Tipos', ylabel = 'Percentual');
plt.show()
```
